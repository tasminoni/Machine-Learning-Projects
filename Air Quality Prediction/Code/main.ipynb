```python
import pandas as pd

file_path = "AirQualityUCI.csv"

data = pd.read_csv(file_path, delimiter=";")

```

```python
data.head(50)
```

```python
data.isnull().sum()
```

```python
# Drop unnamed columns
data = data.loc[:, ~data.columns.str.contains('Unnamed')]


# Reset the index
data.reset_index(drop=True, inplace=True)

```

```python
data.head(50)
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
# Check for duplicates
print(data.duplicated().sum())

#Remove duplicates if needed
data = data.drop_duplicates()
```

```python
print(data.duplicated().sum())
```

```python
data.dtypes
```

```python
# Replace commas with full stops (decimal conversion) in the relevant columns
columns_to_fix = ['CO(GT)', 'C6H6(GT)', 'T', 'RH', 'AH']

# Convert the values in these columns to float, replacing commas with dots
for column in columns_to_fix:
    data[column] = data[column].astype(str).str.replace(',', '.').astype(float)
```

```python
data.head(50)
```

```python
data.dtypes
```

```python
# Drop a specific column by name
data = data.drop(["Date","Time"], axis=1)
```

```python
data.info()
```

```python
# Replace -200 with 0 in the entire dataset
data.replace(-200, 0, inplace=True)

```

```python
data.head(50)
```

```python
data.isnull().sum()
```

```python
data
```

```python
# Remove rows with all NaN values
data.dropna(how='all', inplace=True)
```

```python
data.shape
```

```python
print(data.duplicated().sum())
```

```python
# Create another CSV file and copy the clean data into it
new_file_path = "AirQualityUCI_copy.csv"
data.to_csv(new_file_path, index=False)

new_file_path
```
